```python
PIZZA = 1
STUDENT = 2
EMPTY = 0

edges = []
vertices = []
adjacents = {}
node_role = {}
m_edge = 0
n_lagh_edge = 0
lagh_edges = {}
n_student = 0
students_node = []
pizzas_node = []
goal = {}
priority_against = {}
```

```python
class State:
    def __init__(self, node_number, parent_node, unsatisfied_students, unhandeled_priorities, holded_pizza, lagh_edges, node_role, passed_time):
      self.node_number = node_number
      self.parent_node = parent_node
      self.unsatisfied_students = unsatisfied_students
      self.unhandeled_priorities = unhandeled_priorities
      self.holded_pizza = holded_pizza
      self.passed_time = passed_time
      self.lagh_edges = lagh_edges
      self.node_role = node_role
    def __str__(self):
      return f"{self.node_number}({self.unsatisfied_students},{self.unhandeled_priorities},{self.holded_pizza})"
```

```python
def printPath(last_state):
  if(last_state == None):
    return
  printPath(last_state.parent_node)
  print("-> state", last_state.node_number,"till second", last_state.passed_time, ) 
```

```python
def read_file(file_name):
  f = open(file_name, "r")

  n_vertice_m_edge = f.readline().split(' ')
  m_edge = int(n_vertice_m_edge[1])

  for i in range(m_edge):
      u_v = f.readline().split(' ')
      u = int(u_v[0])
      v = int(u_v[1])
      this_edge = frozenset({u, v})
      edges.append(this_edge)
      if(adjacents.get(u) == None):
        adjacents[u] = []
      adjacents[u].append(v)
      if(adjacents.get(v) == None):
        adjacents[v] = []
      adjacents[v].append(u)
      vertices.append(u)
      vertices.append(v)

  vertices2 = set(vertices)
  for i in vertices2:
    node_role[i] = EMPTY   

  n_lagh_edge = int(f.readline())
  for i in range(n_lagh_edge):
      index_time = f.readline().split(' ')
      index = int(index_time[0])
      latancy = int(index_time[1])
      lagh_edges[edges[index - 1]] =  (latancy, 0)
  
  zirej_curr_node = int(f.readline())
  

  n_student = int(f.readline())
  for i in range(n_student):
      student_information = f.readline().split(' ')
      student_place, pizza_place = [int(x) for x in student_information]
      students_node.append(student_place)
      pizzas_node.append(pizza_place)
      node_role[pizza_place] = PIZZA
      node_role[student_place] = STUDENT
      goal[pizza_place] = student_place

  n_priorities =  int(f.readline())
  for i in range(n_priorities):
      first_stu, second_stu = [students_node[int(x)-1] for x in f.readline().split(' ')]
      priority_against[second_stu] = first_stu

  return zirej_curr_node
```

```python
def BFS(frontier,adjacents):
    goal_states = []
    while(frontier):
        state = frontier.pop(0)
        if(state.unsatisfied_students == [] and state.unhandeled_priorities == {}):
          goal_states.append(state) 
          continue
        explored.add(str(state))
        for neighbor in adjacents[state.node_number]:
            wait_time = 1
            child = start
            unsat_stu = copy.deepcopy(state.unsatisfied_students)
            unhandle_pri = copy.deepcopy(state.unhandeled_priorities)
            holded_pz = copy.deepcopy(state.holded_pizza)
            lagh_edg = copy.deepcopy(state.lagh_edges)
            node_rl = copy.deepcopy(state.node_role)
            curr_edge = frozenset({state.node_number, neighbor})
            l_a = lagh_edg.get(curr_edge)
            if( l_a != None):
              latancy, available_after = l_a
              if(available_after == 0): 
                lagh_edg[curr_edge] = (latancy, state.passed_time + latancy + 1)
              else:
                if (available_after - state.passed_time) > 0:
                  wait_time += (available_after - state.passed_time)
              
            if(node_rl[neighbor] == PIZZA):
              if(state.holded_pizza == -1 and state.unhandeled_priorities.get(goal[neighbor]) == None):
                holded_pz = neighbor
                node_rl[neighbor] = EMPTY
                
            elif(node_rl[neighbor] == STUDENT):
              if(goal.get(state.holded_pizza) == neighbor):
                if neighbor in unsat_stu:
                  unsat_stu.remove(neighbor)
                if(neighbor in state.unhandeled_priorities.values()):
                  key = {i for i in state.unhandeled_priorities if state.unhandeled_priorities[i]==neighbor}
                  for k in key:
                    del unhandle_pri[k]
                holded_pz = -1
                node_rl[neighbor] = EMPTY
            
              
            child = State(neighbor, state, unsat_stu, unhandle_pri, holded_pz, lagh_edg, node_rl, state.passed_time + wait_time)
            if(str(child) not in explored and str(child) not in frontier_str):
              frontier.append(child)
              frontier_str.add(str(child))

    return goal_states
```

```python
def IDS(init):
    for dl in range(100):
        explored.clear()
        states.clear()
        if(DFS(init, dl)):
            return 1  
    return 0

def DFS(state, dl):
    sstr = str(state)
    explored.add(sstr)
    explored_dict[sstr] = dl
    if(state.unsatisfied_students == [] and state.unhandeled_priorities == {}):
        printPath(state)
        return state

    if dl == 0: 
        return 0 
    
    for neighbor in adjacents[state.node_number]:
        wait_time = 1
        child = state
        unsat_stu = copy.deepcopy(state.unsatisfied_students)
        unhandle_pri = copy.deepcopy(state.unhandeled_priorities)
        holded_pz = copy.deepcopy(state.holded_pizza)
        lagh_edg = copy.deepcopy(state.lagh_edges)
        node_rl = copy.deepcopy(state.node_role)
        curr_edge = frozenset({state.node_number, neighbor})
        l_a = lagh_edg.get(curr_edge)
        if( l_a != None):
            latancy, available_after = l_a
            if(available_after == 0): 
                lagh_edg[curr_edge] = (latancy, state.passed_time + latancy + 1)
            else:
                if (available_after - state.passed_time) > 0:
                    wait_time += (available_after - state.passed_time)
        
        if(node_rl[neighbor] == PIZZA):
            if(state.holded_pizza == -1 and state.unhandeled_priorities.get(goal[neighbor]) == None):
                holded_pz = neighbor
                node_rl[neighbor] = EMPTY
            
        elif(node_rl[neighbor] == STUDENT):
            if(goal.get(state.holded_pizza) == neighbor):
                if neighbor in unsat_stu:
                    unsat_stu.remove(neighbor)
                if(neighbor in state.unhandeled_priorities.values()):
                    key = {i for i in state.unhandeled_priorities if state.unhandeled_priorities[i]==neighbor}
                    for k in key:
                        del unhandle_pri[k]
                holded_pz = -1
                node_rl[neighbor] = EMPTY
            
        child = State(neighbor, state, unsat_stu, unhandle_pri, holded_pz, lagh_edg, node_rl, state.passed_time + wait_time)
        child_str = str(child)
        if not child_str in explored or (child_str in explored and explored_dict[child_str] < dl-1):
            states.append(child)
            explored_dict[child_str] = dl-1
            if DFS(child, dl-1):
                return 1

    return 0
```

```python
class MinHeap:
    def __init__(self):
        self.heap_list = [0]
        self.current_size = 0
 
    def sift_up(self, i):
        while i // 2 > 0:
            if self.heap_list[i].f < self.heap_list[i // 2].f:
                self.heap_list[i], self.heap_list[i // 2] = self.heap_list[i // 2], self.heap_list[i]
            i = i // 2
 
    def insert(self, k):
        self.heap_list.append(k)
        self.current_size += 1
        self.sift_up(self.current_size)

    def min_child(self, i):
        if (i * 2)+1 > self.current_size:
            return i * 2
        else:
            if self.heap_list[i*2].f < self.heap_list[(i*2)+1].f:
                return i * 2
            else:
                return (i * 2) + 1
 
    def sift_down(self, i):
        while (i * 2) <= self.current_size:
            mc = self.min_child(i)
            if self.heap_list[i].f > self.heap_list[mc].f:
                self.heap_list[i], self.heap_list[mc] = self.heap_list[mc], self.heap_list[i]
            i = mc
 
    def delete_min(self):
        if len(self.heap_list) == 1:
            return 'Empty heap'
        root = self.heap_list[1]
        self.heap_list[1] = self.heap_list[self.current_size]
        *self.heap_list, _ = self.heap_list
        self.current_size -= 1
        self.sift_down(1)
        return root

frontier = MinHeap()
```

```python
class State:
    def __init__(self, node_number, parent_node, unsatisfied_students, unhandeled_priorities, holded_pizza, lagh_edges, node_role, estimated_cost, g, f, passed_time):
      self.node_number = node_number
      self.parent_node = parent_node
      self.unsatisfied_students = unsatisfied_students
      self.unhandeled_priorities = unhandeled_priorities
      self.holded_pizza = holded_pizza
      self.passed_time = passed_time
      self.lagh_edges = lagh_edges
      self.node_role = node_role
      self.estimated_cost = estimated_cost
      self.g = g
      self.f = f
    def __str__(self):
      return f"{self.node_number}({self.unsatisfied_students},{self.unhandeled_priorities},{self.holded_pizza})"
```

```python
def estimate_cost(state):
    cost = 2 * len(state.unsatisfied_students)
    if goal.get(state.holded_pizza) and goal.get(state.holded_pizza) not in state.unhandeled_priorities:
        cost -= 1
    return cost
```

```python
def A_STAR(frontier):
    goal_states = []
    while(len(frontier.heap_list) > 1):
        state = frontier.delete_min()
        explored.add(str(state))
        for neighbor in adjacents[state.node_number]:
            wait_time = 1
            child = start
            unsat_stu = copy.deepcopy(state.unsatisfied_students)
            unhandle_pri = copy.deepcopy(state.unhandeled_priorities)
            holded_pz = copy.deepcopy(state.holded_pizza)
            lagh_edg = copy.deepcopy(state.lagh_edges)
            node_rl = copy.deepcopy(state.node_role)
            curr_edge = frozenset({state.node_number, neighbor})
            l_a = lagh_edg.get(curr_edge)
            if( l_a != None):
              latancy, available_after = l_a
              if(available_after == 0): 
                lagh_edg[curr_edge] = (latancy, state.passed_time + latancy + 1)
              else:
                if (available_after - state.passed_time) > 0:
                  wait_time += (available_after - state.passed_time)
            
            if(node_rl[neighbor] == PIZZA):
              if(state.holded_pizza == -1 and state.unhandeled_priorities.get(goal[neighbor]) == None):
                holded_pz = neighbor
                node_rl[neighbor] = EMPTY
                
            elif(node_rl[neighbor] == STUDENT):
              if(goal.get(state.holded_pizza) == neighbor):
                if neighbor in unsat_stu:
                  unsat_stu.remove(neighbor)
                if(neighbor in state.unhandeled_priorities.values()):
                  key = {i for i in state.unhandeled_priorities if state.unhandeled_priorities[i]==neighbor}
                  for k in key:
                    del unhandle_pri[k]
                holded_pz = -1
                node_rl[neighbor] = EMPTY
            
            child = State(neighbor, state, unsat_stu, unhandle_pri, holded_pz, lagh_edg, node_rl, 0, 0, 0, state.passed_time + wait_time)
            
            if(child.unsatisfied_students == [] and child.unhandeled_priorities == {}):
                goal_states.append(child)
                continue
            
            if str(child) in explored:
                continue

            child.estimated_cost = estimate_cost(child)
            child.g = state.g + 1
            child.f = child.g + child.estimated_cost

            if str(child) in frontier_str:
                if child.g <= frontier_dict[str(child)]:
                    frontier_dict[str(child)] = child.g
                    frontier.insert(child)
                    frontier_str.add(str(child))
                continue
            
            frontier.insert(child)
            frontier_str.add(str(child))
            frontier_dict[str(child)] = state.g       

    return goal_states
```
